```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import tifffile as tiff
import keras_tuner as kt
import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.densenet import DenseNet121, preprocess_input
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
from sklearn.model_selection import StratifiedKFold, train_test_split
from tensorflow.keras.metrics import AUC
from tensorflow.keras.models import load_model
from IPython.display import Image, display
import matplotlib as mpl
```

```python
#defining the directories
#main directory
data_dir = "..\\"

#test directory
test_ds = os.path.join(data_dir, "test")
test_path = os.path.join(data_dir, "sample_submission.csv")
kaggle_test_df = pd.read_csv(test_path)

#training directory
train_ds = os.path.join(data_dir, "train")
labels_path = os.path.join(data_dir, "train_labels.csv")
labels_df = pd.read_csv(labels_path)
count = labels_df['label'].value_counts()
print(count)

class_names = os.listdir(data_dir)
print(class_names)
print('\n')
```

```python
#kilder
#https://www.datacamp.com/tutorial/python-pie-chart
#plotting a pie chart to show data distribution
categories = ["No cancer", "Cancer"]
plt.pie(count, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title("The data balance")
plt.show()
```

```python
#kilde
#stackoverflow.com/questions/61880368/apply-os-path-join-method-within-a-dataframe
#https://www.kaggle.com/code/vernondsouza123/beginner-transfer-learning
#https://stackoverflow.com/questions/19471814/display-multiple-images-in-one-ipython-notebook-cell

#adding a path column to the labels dataframe so we have easy access to the images
labels_df["path"] = labels_df["id"].apply(lambda x: os.path.join(train_ds, x + ".tif"))

#making a small sample for the plotting
positive_labels = labels_df[labels_df['label'] == 1].sample(5, random_state=42)
negative_labels = labels_df[labels_df['label'] == 0].sample(5, random_state=42)

fig, axes = plt.subplots(2, 5, figsize=(15, 6))
#plotting the cancer positive and negative images 
for ax, row_data in zip(axes[0], positive_labels.itertuples()):
    img = tiff.imread(row_data.path)
    ax.imshow(img)
    ax.set_title(categories[1])
    ax.axis("off")

for ax, row_data in zip(axes[1], negative_labels.itertuples()):
    img = tiff.imread(row_data.path)
    ax.imshow(img)
    ax.set_title(categories[0])
    ax.axis("off")
plt.show()
```

```python
#defining an equal amount of data to be used and concating it into the same dataframe
positive_labels = labels_df[labels_df['label'] == 1].sample(10000, random_state=42)
negative_labels = labels_df[labels_df['label'] == 0].sample(10000, random_state=42)
subset_df = pd.concat([positive_labels, negative_labels]).sample(frac=1, random_state=42).reset_index(drop=True)

print(subset_df["label"].value_counts())

subset_df.head()
```

```python
#kilder
#https://medium.com/the-owl/k-fold-cross-validation-in-keras-3ec4a3a00538

#defining some constants for the pipeline
#increased the resolution to 128x128 since DenseNet121 usually takes 224x224 but
#this was a middle point between hardware limitation and model needs
IMG_SIZE   = 128 
BATCH_SIZE = 64

#augmetation and normalization logic for the training
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input, #using preprocess_input since DenseNet121 has its own normalization function
    horizontal_flip=True,
    zoom_range=0.2,
    fill_mode='nearest',
    width_shift_range=0.1,
    height_shift_range=0.1,
    brightness_range=[0.8, 1.2],
)

#normalization logic for both the validation and testing
minimal_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

#splitting the training and the "rest_df" which will be split again to the validation and test sets
#the current split is 70% test, 15% validation, 15% testing 
train_df, rest_df = train_test_split(subset_df, test_size=0.30, stratify=subset_df["label"], random_state=42)
val_df, test_df = train_test_split(rest_df, test_size=0.5, stratify=rest_df["label"], random_state=42)

print("Train:", len(train_df), "Val:", len(val_df), "Test:", len(test_df))
```

```python
#kilder
#https://medium.com/the-owl/k-fold-cross-validation-in-keras-3ec4a3a00538

#defining the data characteristics/pipeline
#wrapped into a function since its used both in hyperparameter search and stratified kfold 
def data_generation(training_data, validation_data):
    
    training_gen = train_datagen.flow_from_dataframe(
    dataframe = training_data,
    directory = None,
    x_col = "path",
    y_col = "label",
    target_size = (IMG_SIZE, IMG_SIZE),
    color_mode = "rgb",
    class_mode = "raw",
    batch_size = BATCH_SIZE,
    seed = 42,
    shuffle = True
)
    validation_gen = minimal_datagen.flow_from_dataframe(
    dataframe = validation_data,
    directory = None,
    x_col = "path",
    y_col = "label",
    target_size = (IMG_SIZE, IMG_SIZE),
    color_mode = "rgb",
    class_mode = "raw",
    batch_size = BATCH_SIZE,
    shuffle = False
)  
    return training_gen, validation_gen

train_gen, val_gen = data_generation(train_df, val_df)
print(train_gen.batch_size)
```

```python
#kilder
#https://medium.com/the-owl/k-fold-cross-validation-in-keras-3ec4a3a00538

#preparing the testing for both kaggle and the custom test on validation later on the end
kaggle_test_df["path"] = kaggle_test_df["id"].apply(lambda x: os.path.join(test_ds, x + ".tif"))

print(kaggle_test_df["path"].head())

kaggle_test_gen = minimal_datagen.flow_from_dataframe(
    dataframe=kaggle_test_df,
    directory=None,
    x_col="path",
    y_col=None,              
    target_size=(IMG_SIZE, IMG_SIZE),
    color_mode="rgb",
    class_mode=None,         
    batch_size=BATCH_SIZE,
    shuffle=False            
)

test_gen = minimal_datagen.flow_from_dataframe(
    dataframe   = test_df,
    directory   = None,
    x_col       = "path",
    y_col       = "label",
    target_size = (IMG_SIZE, IMG_SIZE),
    color_mode  = "rgb",
    class_mode  = "raw",
    batch_size  = BATCH_SIZE,
    shuffle     = False,
)
```

```python
#kilder
#https://jatit.org/volumes/Vol102No16/4Vol102No16.pdf

#the architecture is a transfer learning cnn model with densenet121 as the backbone which is partually unfrozen
#i also used VGG19 and EfficientNetV2B0 aswell but this was a good middle place betwen preformance and the heavy load on my machine

def transfer_model(hp):
    base_model = DenseNet121(#chose densenet because of its proven preformance on the research paper above
        weights="imagenet",
        include_top=False,
        input_shape=(IMG_SIZE, IMG_SIZE, 3),
    )

    for layer in base_model.layers[:-8]:#freezing every layer except for the last 8th layer since my machine couldnt handle much more
        layer.trainable = False

    x = base_model.output
    x = layers.GlobalAveragePooling2D()(x)
    
    x = Dense(units = hp.Int('units', min_value = 256, max_value = 512, step=128), activation='relu')(x)
    #added BatchNormalization to make the training process smoother for the model 
    #(i added it while working with a bigger modedl:VGG19 but it was still great to use here)
    x = BatchNormalization()(x)
    x = Dropout(hp.Float('dropout', 0.3, 0.5, step=0.1))(x)#introduced a good range for dropout rate to reduce overfitting

    outputs = layers.Dense(1, activation='sigmoid')(x)# Sigmoid output since its binary classification
    model = models.Model(base_model.input, outputs)

    hp_learning_rate = hp.Choice('learning_rate', values=[1e-4, 3e-5, 1e-5])

    model.compile(#Compiling the model with the learning rate, adding binary_crossentropy while also focusing on accuracy and auc as metrics
        optimizer=optimizers.Adam(learning_rate=hp_learning_rate),
        loss='binary_crossentropy',
        metrics=['accuracy', AUC(name='auc')]#auc is important for the kaggle submission and will be our main metric for early stopping
    )
    return model
```

```python
#kilder
#https://www.tensorflow.org/tutorials/keras/keras_tuner
#https://www.jatit.org/volumes/Vol102No16/4Vol102No16.pdf
#https://keras.io/api/applications/densenet/

#hyperparameter tuning going through all the configuration while focusing on the ones with the most val_auc 
def transfer_hypermodel(hp):
    return transfer_model(hp)

tuner = kt.Hyperband(
    transfer_hypermodel,
    objective=kt.Objective('val_auc', direction='max'),
    max_epochs=15,
    factor=3,
    directory='tuner',
    project_name='densenet-121',
    #overwrite=True
)

#gave early stopping a bigger patience value since i noticed it getting better after more than 3 epoches
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor="val_auc",
    patience=5,
    restore_best_weights=True
)

tuner.search(
    train_gen,
    validation_data=val_gen,
    epochs=15,
    callbacks=[early_stop],
)

#saving and outputting the best hyperparameters
best_hp_transfer = tuner.get_best_hyperparameters(1)[0]

print(f"""
Best hyperparameters:
- Units: {best_hp_transfer.get('units')}
- Learning rate: {best_hp_transfer.get('learning_rate')}
- Dropout: {best_hp_transfer.get('dropout')}
""")
```

```python
#kilder
#https://www.tensorflow.org/tutorials/keras/keras_tuner

#training the hypermodel with the best hyperparameters with the goal of getting the best epoch to reduce overfitting
model = transfer_model(best_hp_transfer)

history_best = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=20,
    callbacks=[early_stop]
)
#the main metric we look at is AUC but i wanted to still track accuracy just incase there was a huge difference
#both metrics usually reach their highest position within 2-3 epochs of eachother 
val_acc_per_epoch = history_best.history["val_accuracy"]
best_epoch_transfer = int(np.argmax(val_acc_per_epoch)) + 1 
best_val_acc = float(np.max(val_acc_per_epoch))

val_auc_per_epoch = history_best.history["val_auc"]
best_auc_epoch = int(np.argmax(val_auc_per_epoch)) + 1
best_val_auc = float(np.max(val_auc_per_epoch))

print(f"Best epoch for accuracy: {best_epoch_transfer}")
print(f"Best val accuracy: {best_val_acc:.4f}")

print(f"Best epoch for AUC: {best_auc_epoch}")
print(f"Best val AUC: {best_val_auc:.4f}")
```

```python
#kilder
#https://www.tensorflow.org/tutorials/keras/keras_tuner

#rebuilding the hypermodel with the best hyperparameters and setting the best epoch count
#this is the model that will be further used for some of the figures below and in the Stratified kfold
final_model = transfer_model(best_hp_transfer)

history_final = final_model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=best_auc_epoch,
    verbose=1,
)
```

```python
#kilder
# https://www.analyticsvidhya.com/blog/2021/09/how-to-apply-k-fold-averaging-on-deep-learning-classifier/
# https://medium.com/the-owl/k-fold-cross-validation-in-keras-3ec4a3a00538

#stratified kfold logic wrapped in a function so that it can be called incase there was multiple models to test
#the parameters taking in both a model and amount of epochs
def skf(model_choice, epochs):
    X = subset_df.path
    y = subset_df.label
    rounds = 5
    fold = 1

    skf = StratifiedKFold(n_splits=rounds, shuffle=True, random_state=42)
    fold_acc = []
    fold_loss = []
    fold_auc = []
    
    for train_idx, val_idx in list(skf.split(X, y)):
        
        #a numpy array that has the indexes for the training and valiadtion data used in every fold
        training_fold = subset_df.iloc[train_idx].copy()
        validation_fold   = subset_df.iloc[val_idx].copy()

        #making new generators that can refer to the correct dataframe with the indexies of the current training and validation set
        train_fold_gen, val_fold_gen = data_generation(training_fold, validation_fold)
        
        print("\nFold", fold, "/", rounds)

        model = model_choice()

        model.fit(
            train_fold_gen,
            validation_data=val_fold_gen,
            epochs=epochs,
            callbacks=[early_stop],
            verbose=1,
        )

        #evaluating the model on the validation set 
        results = model.evaluate(val_fold_gen, verbose=0)
        results = dict(zip(model.metrics_names,results))

        val_auc = results['auc']
        val_accuracy = results['accuracy']
        val_loss = results['loss']

        print("\nFold", fold, "val_acc:", val_accuracy, "|| val_loss:", val_loss)

        #adding the main metrics into the aray to later get a mean value on all 5 folds
        fold = fold + 1
        fold_auc.append(val_auc)
        fold_acc.append(val_accuracy)
        fold_loss.append(val_loss)

        tf.keras.backend.clear_session()
        
    print("\nCV mean validation AUC:", np.mean(fold_auc), "|| Std", np.std(fold_auc))
    print("\nCV mean validation accuracy:", np.mean(fold_acc), "|| Std", np.std(fold_acc))
    print("\nCV mean validation loss:", np.mean(fold_loss), "|| Std", np.std(fold_loss))
    
    return fold_auc, fold_acc , fold_loss
```

```python
#kilde
#https://medium.com/the-owl/k-fold-cross-validation-in-keras-3ec4a3a00538

#a helper function that can call the hypermodel and the skf function with the parameters of the model choice and best epoch
def build_transfer_model():
    return transfer_model(best_hp_transfer)

transfer_model_evalutation = skf(model_choice=build_transfer_model, epochs=best_auc_epoch)
```

```python
#kilder
#https://keras.io/examples/vision/grad_cam/
#https://github.com/Bsingstad/Data-Mining-and-Analytics-2025/blob/main/Forelesning201025%20del%202/04%20CNN-hund-katt.ipynb

#implementing GradCam to see into the blackbox of the model
#this helps us see what the model is looking at in the images and where its finding the cancer cells

def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    # First, we create a model that maps the input image to the activations
    # of the last conv layer as well as the output predictions
    grad_model = keras.models.Model(
        model.inputs, [model.get_layer(last_conv_layer_name).output, model.output]
    )

    # i compute the gradient of the top predicted class for our input image
    # with respect to the activations of the last conv layer
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]

    # This is the gradient of the output neuron (top predicted or chosen)
    # with regard to the output feature map of the last conv layer
    grads = tape.gradient(class_channel, last_conv_layer_output)

    # This is a vector where each entry is the mean intensity of the gradient
    # over a specific feature map channel
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    # We multiply each channel in the feature map array
    # by "how important this channel is" with regard to the top predicted class
    # then sum all the channels to obtain the heatmap class activation
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    # For visualization purpose, i will also normalize the heatmap between 0 & 1
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()
```

```python
#kilder
#https://keras.io/examples/vision/grad_cam/
#https://github.com/Bsingstad/Data-Mining-and-Analytics-2025/blob/main/Forelesning201025%20del%202/04%20CNN-hund-katt.ipynb

def save_and_display_gradcam(img, heatmap, cam_path="cam.jpg", alpha=0.8):
    # Load the original image
    img = keras.utils.load_img(img_path)
    img = keras.utils.img_to_array(img)
    
    # Rescale heatmap to a range 0-255
    heatmap = np.uint8(255 * heatmap)

    # Use jet colormap to colorize heatmap
    jet = mpl.colormaps["jet"]

    # Use RGB values of the colormap
    jet_colors = jet(np.arange(256))[:, :3]
    jet_heatmap = jet_colors[heatmap]

    # Create an image with RGB colorized heatmap
    jet_heatmap = tf.keras.utils.array_to_img(jet_heatmap)
    jet_heatmap = jet_heatmap.resize((img.shape[1], img.shape[0]))
    jet_heatmap = tf.keras.utils.img_to_array(jet_heatmap)

    # Superimpose the heatmap on original image
    superimposed_img = jet_heatmap * alpha + img
    
    superimposed_img = tf.keras.utils.array_to_img(superimposed_img)
    plt.imshow(superimposed_img)
```

```python
#kilder
#https://github.com/Bsingstad/Data-Mining-and-Analytics-2025/blob/main/Forelesning201025%20del%202/04%20CNN-hund-katt.ipynb
#https://keras.io/examples/vision/grad_cam/
#
index = 12
amount = 4

for img_indx in range(index,index+amount):
    #using img_indx to go through the indexes from the validation  
    img_path = val_df['path'].iloc[img_indx]
    #then getting the label type for the plotting later
    img_id = val_df['label'].iloc[img_indx]

    img = tf.keras.utils.load_img(img_path, target_size=(IMG_SIZE,IMG_SIZE))
    img_array = tf.keras.utils.img_to_array(img)

    img_array = img_array / 255.0
    img_array = np.expand_dims(img_array, axis=0) 

    #after getting all the proper data: the image array, the tuned hypermodel and the last convolutional layer are added
    heat_map = make_gradcam_heatmap(img_array, final_model, "conv5_block16_2_conv", pred_index=None)
    
    plt.figure(figsize=(4,3))
    plt.imshow( tf.keras.utils.array_to_img(img))
    plt.title(f"{categories[img_id]} - Original")
    plt.axis("off")
    
    plt.figure(figsize=(4,3))
    save_and_display_gradcam(img_path,heat_map)
    plt.title(f"{categories[img_id]} - GradCAM")
    plt.axis("off")
    plt.show()
```
